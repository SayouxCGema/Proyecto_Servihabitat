# Proyecto de ingenieria de datos web Servihabitat

- El objetivo de este proyecto, es extraer una base de datos con información de viviendas en venta en la ciudad de Barcelona con sus respectivos precios. Posteriormeente, la base de datos, puede ser utlizada para realizar análisis de precios del mercado. La estracción de los datos se realiza en un total de 11 páginas. Fecha agosto 2024.

```python
# librerias para la extracción de datos

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import undetected_chromedriver as uc


import pandas as pd
import requests
```

```python
# acceso version navegacion usuario a la web de servihabitat

spy = uc.Chrome()
```

```python
# para que no detecta el scrapper

spy.implicitly_wait(10)

# asignacion de pagina de busqueda
html = spy.page_source

# almacenamiento de la url
url = 'https://www.servihabitat.com/es/venta/vivienda/barcelona'

# pedir que el espia acceda a la web

spy.get(url)

# pedir ver la web en html

spy_soup = BeautifulSoup(html,'html.parser')
```

```python
# Creación de listas vacías
precio = []
detalles = []
m2 = []
hab = []
baños = []
```

```python
for span in spy_soup.find_all('span', class_='price', id='price'):
    precio.append(span.text.strip())
    

for h2 in spy_soup.find_all('h2', id='description', class_='features-address'):
    detalles.append(h2.text.strip())
    
    
for li1 in spy_soup.find_all('li', id='superficie', class_='features-list-metros'):
    m2.append(li1.text.strip())

for li in spy_soup.find_all('li',id='numero-hab', class_='features-list-habitaciones'):
    hab.append(li.text.strip())
    
    
for li in spy_soup.find_all('li', id='numero-ban',class_='features-list-banos'):
    baños.append(li.text.strip())
```
